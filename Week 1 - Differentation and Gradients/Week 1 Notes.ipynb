A single value -> Scalar - > A tensor with no dimensions
A vector -> the values are represented in a single dimensions
A tensor is a generic data structure

In the context of data science, tensors are multi-dimensional arrays of numbers that represent complex data. They are the fundamental data structures used in machine learning and deep learning frameworks like TensorFlow and PyTorch.

Tensors are particularly useful for handling complex data such as images, audio, and text. For example, a color image can be represented as a 3D tensor with dimensions corresponding to height, width, and color channels (red, green, and blue). Similarly, a video can be represented as a 4D tensor with time being the additional dimension.
